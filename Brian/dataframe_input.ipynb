```python
def historical_prices(weights_df):
    if isinstance(weights_df,pd.DataFrame):
        tickers_list=list(weights_df.index)
        dataframe_list=[]
        api_calls_made=0
        for ticker in tickers_list:
            if api_calls_made<10:
                dataframe_list.append(yahoo_api_call(ticker))
                api_calls_made+=1
            else:
                time.sleep(100)
                dataframe_list.append(yahoo_api_call(ticker))
                api_calls_made=1
        df_historical=pd.concat(dataframe_list,axis=1,keys=tickers_list)
        df_historical.dropna(how='any',inplace=True)
        df_historical=df_historical[~(df_historical==0).any(axis=1)]
        return(df_historical)
    else:
        print("Please input a dataframe into the historical_prices function")
```

```python
dataframe_output=historical_prices(test_dataframe)
```
