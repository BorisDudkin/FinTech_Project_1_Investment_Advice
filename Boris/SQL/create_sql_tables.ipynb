```python
# Import modules
import pandas as pd
import sqlalchemy
from pathlib import Path
```

```python
# read capacity questionnaires into a list of dataframes risk_capacity:
my_questionnaires=['Q1_Capacity.csv','Q2_Capacity.csv','Q3_Capacity.csv','Q4_Capacity.csv','Q5_Capacity.csv']
risk_capacity=[]

for quest in my_questionnaires:
    csvpath = Path("../Resources/"+quest)
    pd_name= quest+"_df"
    pd_name = pd.read_csv(csvpath)
    risk_capacity.append(pd_name)
    
display(risk_capacity)
# read tolerance questionnaire into a dataframes risk_tolerance:
risk_tolerance_df = pd.read_csv(csvpath)
csvpath = Path("../Resources/Q_Tolerance.csv")
risk_tolerance_df = pd.read_csv(csvpath)
display(risk_tolerance_df)
read portfolios into a portfolios_df dataframe:
csvpath = Path("../Resources/Portfolios.csv")
portfolios_df = pd.read_csv(csvpath)
display(portfolios_df)
# read sectors into a sectors_mapping_df dataframe:
csvpath = Path("../Resources/Sector_mapping.csv")
sectors_mapping_df = pd.read_csv(csvpath)
display(sectors_mapping_df)
```

## Create the SQLite database and engine

```python
# Create a temporary sqlite database
database_connection_string = 'sqlite:///../Resources/investor.db'
```

```python
# Database connection object
engine = sqlalchemy.create_engine(database_connection_string)
engine
```

## Create a Table From a Pandas DataFrame

```python
# Confirm table creation
engine.table_names()
```

```python
# Create a table corresponding to the risk tolerance questions
risk_tolerance_df.to_sql('risk_tolerance_questions', engine, index=False, if_exists='replace')
```

```python
# Confirm table creation
engine.table_names()
```

```python
# # Create tables corresponding to the risk capacity questions
risk_capacity[0].to_sql('Q1_Capacity', engine, index=False, if_exists='replace')
risk_capacity[1].to_sql('Q2_Capacity', engine, index=False, if_exists='replace')
risk_capacity[2].to_sql('Q3_Capacity', engine, index=False, if_exists='replace')
risk_capacity[3].to_sql('Q4_Capacity', engine, index=False, if_exists='replace')
risk_capacity[4].to_sql('Q5_Capacity', engine, index=False, if_exists='replace')
```

```python
# # Create a table holding  the sector/industry/market cap/name mapping
sectors_mapping_df.to_sql('sector_mapping', engine, index=False, if_exists='replace')
```

```python
# Confirm table creation
engine.table_names()
```

```python
# Create a portfolios table
portfolios_df.to_sql('portfolios', engine, index=False, if_exists='replace')
```

```python
engine.table_names()
```

```python
capacity_tables=['Q1_Capacity','Q2_Capacity','Q3_Capacity','Q4_Capacity','Q5_Capacity']
risk_capacity=[]

for table in capacity_tables:
    pd_name= table+"_df"
    pd_name = pd.read_sql_table(table, con=engine)
    pd_name=pd_name.set_index('Questions')
    risk_capacity.append(pd_name)
```
